```python
from transfermarkt import Player
```

```python
playerId = 28003

messi = Player(playerId)
```

```python
messi.matches()
```

```python
messi.performance(by='competition') # Messi stats by competition
```

```python
messi.performance(by='position') # Messi stats by position
```

```python
messi.matches(competition='LaLiga') # Messi all in LaLiga
```

```python
messi.matches(opponent='Real Madrid') # Messi all matches against Real Madrid
```

```python
messi.matches(opponent='Real Madrid', competition='UEFA Champions League') # Messi all matches against Real Madrid in UEFA Champions League
```

```python
messi.matches()['goals'].sum() # Messi all-time goals
```

```python
messi.matches()['assists'].sum() # Messi all-time goals
```
