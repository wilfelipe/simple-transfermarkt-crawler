```python
from transfermarkt import Player
```

```python
playerId = 28003

messi = Player(playerId)
```

```python
messi.matches()
```

```python
messi.matches(date = '2015').head() # All messi matches in 2015
```

```python
messi.matches(competition='LaLiga').head() # Messi all in LaLiga
```

```python
messi.matches(opponent='Real Madrid', competition='UEFA Champions League').head() # Messi all matches against Real Madrid in UEFA Champions League
```

```python
messi.performance(by='competition') # Messi stats by competition
```

```python
messi.performance(by='position') # Messi stats by position
```

```python
messi.matches()['goals'].sum() # Messi all-time goals
```

```python
messi.matches()['assists'].sum() # Messi all-time goals
```
